```python
def split_image(img_path):
    img = Image.open(img_path)
    min_size = min(img.size[0], img.size[1])
    half_size = min_size // 2
    img = img.resize((min_size, min_size))
    left_upper = (0, 0, half_size, half_size)
    left_lower = (0, half_size, half_size, min_size)
    right_upper = (half_size, 0, min_size, half_size)
    right_lower = (half_size, half_size, min_size, min_size)
    return (img.crop(left_upper), img.crop(left_lower), img.crop(right_upper), img.crop(right_lower))
```

```python
left_upper, left_lower, right_upper, right_lower = split_image('python.png')
```
