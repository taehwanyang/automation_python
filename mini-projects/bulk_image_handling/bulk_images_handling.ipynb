### 한장의 이미지로 30% ~ 60% 리사이징한 이미지 100장 만들기

```python
from PIL import Image
import numpy as np
import os
```

### 최소, 최대 비율 정의

```python
low=10 
high=90
```

### 저장할 폴더 이름 정의

```python
dest_folder = 'bulk_images'
```

```python
img = Image.open('example.jpeg')
```

```python
display(img)
```

```python
if not os.path.exists(dest_folder):
    os.mkdir(dest_folder)
```

```python
for i in range(1,101):
    reduced_width = int(img.width * (np.random.randint(low, high) / 100))
    reduced_height = int(img.height * (np.random.randint(low, high) / 100))
    img_resized = img.resize((reduced_width, reduced_height))
    print(img_resized.size)
    img_resized.save(os.path.join(dest_folder, f'example_{i}.jpeg'))
```
